```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#transfermarkt data cleaning
path = r'C:\\Users\\aaron\\prem_league_analytics\\data\\'
df_tm_performance = pd.read_csv(path+'performance.csv')
```

```python
df_perf16 = df_tm_performance[['name','age', 'height', 'position', 'current club', '15/16 games', '15/16 goals', '15/16 assists', '15/16 minutes']]
df_perf17 = df_tm_performance[['name','age', 'height', 'position', 'current club', '16/17 games', '16/17 goals', '16/17 assists', '16/17 minutes']]
df_perf18 = df_tm_performance[['name','age', 'height', 'position', 'current club', '17/18 games', '17/18 goals', '17/18 assists', '17/18 minutes']]
df_perf19 = df_tm_performance[['name','age', 'height', 'position', 'current club', '18/19 games', '18/19 goals', '18/19 assists', '18/19 minutes']]
df_perf20 = df_tm_performance[['name','age', 'height', 'position', 'current club', '19/20 games', '19/20 goals', '19/20 assists', '19/20 minutes']]
```

```python
path = r'C:\Users\aaron\prem_league_analytics\data\transfermarkt\performance_data\ '

df_perf16.to_csv(path+'perf16.csv')
df_perf17.to_csv(path+'perf17.csv')
df_perf18.to_csv(path+'perf18.csv')
df_perf19.to_csv(path+'perf19.csv')
df_perf20.to_csv(path+'perf20.csv')
```

```python
df_perf16 = pd.read_csv(path+'perf16.csv')
df_perf16['year'] = '15/16'
df_perf17 = pd.read_csv(path+'perf17.csv')
df_perf17['year'] = '16/17'
df_perf18 = pd.read_csv(path+'perf18.csv')
df_perf18['year'] = '17/18'
df_perf19 = pd.read_csv(path+'perf19.csv')
df_perf19['year'] = '18/19'
df_perf20 = pd.read_csv(path+'perf20.csv')
df_perf20['year'] = '19/20'
```

```python
df_perf16.rename(columns={'15/16 games':'games','15/16 goals': 'goals', '15/16 assists' : 'assists', '15/16 minutes': 'minutes'} , inplace=True)
df_perf17.rename(columns={'16/17 games':'games','16/17 goals': 'goals', '16/17 assists' : 'assists', '16/17 minutes': 'minutes'} , inplace=True)
df_perf18.rename(columns={'17/18 games':'games','17/18 goals': 'goals', '17/18 assists' : 'assists', '17/18 minutes': 'minutes'} , inplace=True)
df_perf19.rename(columns={'18/19 games':'games','18/19 goals': 'goals', '18/19 assists' : 'assists', '18/19 minutes': 'minutes'} , inplace=True)
df_perf20.rename(columns={'19/20 games':'games','19/20 goals': 'goals', '19/20 assists' : 'assists', '19/20 minutes': 'minutes'} , inplace=True)
```

```python
df_perf = pd.concat([df_perf16, df_perf17, df_perf18, df_perf19, df_perf20], sort=False)
df_perf
```

```python
df_perf['firstinitial'] = df_perf['name'].astype(str).str[0]
df_perf['lastname'] = df_perf['name'].str.split(' ').str[1]
df_perf['short_name']=df_perf['firstinitial'].astype(str)+'. '+df_perf['lastname']
df_perf
```

```python
df_perf = df_perf.drop(columns=['Unnamed: 0','firstinitial', 'lastname'])
df_perf = df_perf.dropna()
df_perf = df_perf.replace('Arsenal FC U23', 'Arsenal FC')
df_perf
```

```python
order_goals = df_perf.groupby(by=['current club'])['goals'].median().sort_values(ascending=False).index
```

```python
#Historical trends from 
ax=sns.catplot(x="current club", y="goals", kind="box", order = order_goals, data=df_perf)
ax.fig.set_size_inches(25,20)
ax.set_xticklabels(rotation=90)
plt.title('Goals by Club from 2015-2019')
plt.show()
```
